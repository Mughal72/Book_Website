```python
import pandas as pd
import numpy as np
from surprise import Dataset, Reader, KNNBasic
from surprise.model_selection import train_test_split

# Load the book dataset
book_df = pd.read_csv("/content/books.csv", encoding= 'latin-1')

# Load the user dataset
user_df = pd.read_csv("/content/users.csv")

# Prepare the data for Surprise library
reader = Reader(rating_scale=(1, 5))
data = Dataset.load_from_df(user_df[['User-ID', 'ISBN', 'Ratings']], reader)

# Split the data into train and test sets
trainset, testset = train_test_split(data, test_size=0.2, random_state=42)

# Build the collaborative filtering model
model = KNNBasic(k=5, min_k=1)
model.fit(trainset)

# Get book recommendations for each user
for user_id in user_df['User-ID'].unique():
    user_items = user_df.loc[user_df['User-ID'] == user_id, 'ISBN']
    user_unrated_books = [book for book in book_df['ISBN'] if book not in user_items]
    predictions = [model.predict(user_id, book) for book in user_unrated_books]
    recommendations = sorted(predictions, key=lambda x: x.est, reverse=True)[:10]

    # Print the recommended books for the user
    print("Top 10 recommended books for User", user_id)
    for rec in recommendations:
        book_title = book_df.loc[book_df['ISBN'] == rec.iid, 'Book-Title'].values[0]
        print(f"- {book_title} (Estimated rating: {rec.est})")
    print('\n')
```

In this updated code, we iterate over each unique user ID in the user dataset. For each user, we retrieve their preferences (such as previously read books, author, ratings, etc.) from the user dataset.

We then generate book recommendations for that user based on their preferences using the collaborative filtering model. The process remains the same as before, where we make predictions for unrated books, sort the predictions, and print the top 10 recommended books for each user.

This way, the recommendation engine will personalize the recommendations for each user based on their individual preferences.
